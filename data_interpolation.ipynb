```python
import pandas as pd
```

```python
data = pd.read_csv('scarlet_fever_data_china.csv', parse_dates=['day'])
```

```python
data
```

```python
data = data.set_index('day').resample('D').interpolate()
```

```python
data = data[(data['day'] < '2019-01-01')]
```

```python
data
```

```python
data.to_csv('interpolated_scarlet_fever_data_china.csv')
```
